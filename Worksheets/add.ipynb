```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data=pd.read_csv("add.csv")
data
```

```python
x=data[['x','y']]
y=data['sum']
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y, test_size=0.33,random_state=5)
x_train
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
# lINEAR regression object will be created using linear rregression method. if you want to call the fit method to train the model, we need the object of linear regression class.fit method having two parameters, vector and feature matrices. vector is holding the dependent variant and feature matrices hold the independent variant
```

```python
model.score(x_train,y_train)
model.score(x_test,y_test)
```

```python
y_pred=model.predict(x_test)
y_pred
df=pd.DataFrame({'Actual':y_test,'Prediction':y_pred})
df
```
